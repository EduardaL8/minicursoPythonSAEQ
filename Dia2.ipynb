```python
letra = input("Entre com uma letra: ")
letraM = letra.upper()
if letraM == 'F' :
  print ("F-Feminino")
elif letraM == 'M' :
  print("M-Masculino")
else:
  print ("Opção inválida")
```

```python
a = int(input("Entre com o número 1: "))
b = int(input("Entre com o número 2: "))
c = int(input("Entre com o número 3: "))

maior = 0
menor = 0

if a > b and a > c :
  maior = a
elif b > a and b > c :
  maior = b
else:
  maior = c

if a < b and a < c :
  menor = a
elif b < a and b < c :
  menor = b
else:
  menor = c

print (f"O maior número é {maior} e o menor número é {menor}")
```

```python
vetor = []
vetorp = []
vetori = []

for i in range(0, 20):
  n = int(input(f"Digite o número {i}: "))
  vetor.append(n)
  if n%2 == 0 :
    vetorp.append(n)
  else:
    vetori.append(n)
print(vetor)
print(f"o vetor par é {vetorp}")
print(f"o vetor impar é {vetori}")
```

```python
vetor = []
nota = 0
for i in range(0,4):
   n = int(input(f"Digite a nota {i}: "))
   vetor.append(n)
   nota += n

media = nota/len(vetor)
print(f"As notas são {vetor}")
print(f"A média é {media}")
```
